```python
from skimage import io 
# skimage is an opensource python library used in image processing.
from skimage.color import rgb2gray
```

```python
# Writting the code for image processing using python libraries and OOPs

class Edit:
    a=0
    
    # Code for horizontal flip
    def horizontalFlip(self,np_img):
        img_flip=np_img[:,::-1]
        plt.imshow(img_flip)  # imshow() is used to display data as an image i.e on a 2D regular raster
        return img_flip
    
    # Code for vertical flip
    def verticalFlip(self,np_img):
        img_flip=np_img[::-1]
        plt.imshow(img_flip)  
        return img_flip
    
    # Code for inverting colors
    def invertColors(self,np_img):
        img_invert=255-np_img
        plt.imshow(img_invert)
        return img_invert
    
    # Code for grayscaling of image
    def convertGreyscale(self,np_img):
        img_gray=rgb2gray(np_img)
        plt.imshow(img_grey,cmap='gray', vmin=0, vmax=1)
        return img_gray
    
    # Code for converting image to black and white
    def convertBnw(self,np_img):        
        img_bnw = rgb2gray(np_img)
        img_bnw[img_bnw >= 0.85] = 1
        img_bnw[img_bnw < 0.85] = 0
        plt.imshow(img_bnw,cmap = 'gray', vmin = 0,vmax = 1)
        return img_bnw
    
```
